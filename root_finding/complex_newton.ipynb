# Newton method for complex roots

Consider the function
$$
f(x) = x^5 + 1
$$
The roots lie on the unit circle in the complex plane. We can write
$$
-1 = \exp(i(2n+1)\pi), \qquad n=0,1,2,\ldots
$$
The fifth roots are
$$
(-1)^{1/5} = \exp(i(2n+1)\pi/5), \qquad n=0,1,2,3,4
$$

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
from numpy import abs,exp,array,real,imag,pi
import matplotlib.pyplot as plt
```

Lets compute the exact roots and plot them on a graph.

```python
n = array([0,1,2,3,4])
r = exp(1j*(2*n+1)*pi/5) # 5'th roots of -1
plt.plot(real(r),imag(r),'o')
plt.grid(True)
plt.axis('equal')
for x in r:
    print(x,x**5)
```

Define the function and its derivative.

```python
def f(x):
    return x**5 + 1.0

def df(x):
    return 5.0*x**4
```

Now implement Newton method as a function.

```python
def newton(f,df,x0,M=100,eps=1.0e-15):
    x = x0
    for i in range(M):
        dx = - f(x)/df(x)
        x  = x + dx
        print(i,x,abs(f(x)))
        if abs(dx) < eps * abs(x):
            return x
```

We call Newton method with a complex initial guess for the root.

```python
x0 = 1.0 + 1.0j
x = newton(f,df,x0)
print('Root = ',x)
```
